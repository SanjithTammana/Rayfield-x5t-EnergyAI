```python
import pandas as pd
```

```python
file_path = "dataset (2).csv"
df = pd.read_csv(file_path)
```

```python
df
```

```python
data = df.drop(['Biomass Energy'], axis=1)
data
```

```python
def get_quarter(month):
    if 1 <= month <= 3:
        return 1
    elif 4 <= month <= 6:
        return 2
    elif 7 <= month <= 9:
        return 3
    elif 10 <= month <= 12:
        return 4

data["Quarter"] = data["Month"].apply(get_quarter)
data
```

```python
data = data.drop(['Month'], axis=1)
data
```

```python
# 1. Group by Quarter, Sector, and Year
grouped_data = data.groupby(["Quarter", "Sector", "Year"], as_index=False).sum()
data = grouped_data
data
```

```python
data["Sector"] = data["Sector"].replace("Commerical", "Commercial")
data
```

```python
data.to_csv('training_data.csv', index=False)
```
